第一題(1)

```python
n='junyiacademiy'
def reverse(inputn):
	return inputn[len(inputn):0:-1]+inputn[0]

print(n,'==',reverse(n))
```

第一題(2)

```python
m='flipped classroom is important'
def reverse2(inputm):
	a=''
	inputm=inputm.split(' ')
	for i in range(len(inputm)):
		a+=reverse(inputm[i])
		if(i!=len(inputm)-1):
			a+=' '
	return a

print(m,'==',reverse2(m))
```

第二題

```python
n=int(input())
nlist=[]

for i in range(1,n+1):
	if(i%3!=0 and i%5!=0):
		nlist.append(i)
	if(i%15==0):
		nlist.append(i)
print('output:',len(nlist))
```

|---|---|---|---|
|拿出鉛筆| 混合  |不知  | 鉛筆  |  
| 拿出原子筆 | 不知  | 混合  | 原子筆  |   


第一種～拿出的是鉛筆
1.如果從鉛筆袋拿出鉛筆，可以判斷鉛筆袋其實是混合，然後混合是原子筆，原子筆是鉛筆
2.如果從原子筆袋拿出鉛筆，則無法判斷是從鉛筆袋或是混合袋得到
3.如果從混合袋拿出鉛筆，可以判斷鉛筆袋其實是原子筆，然後混合是鉛筆，原子筆是混合

第二種～拿出的是原子筆
1.如果從鉛筆袋拿出原子筆，則無法判斷是從原子筆袋或是混合袋得到
2.如果從原子筆袋拿出原子筆，可以判斷鉛筆袋其實是原子筆，然後混合是鉛筆，原子筆是混合袋
3.如果從混合袋拿出原子筆，可以判斷鉛筆袋其實是混合，然後混合是原子筆，原子筆是鉛筆

第四題
有三個人一起到迪士尼遊玩，中午肚子餓了，去餐廳點了一份現在最夯的冰雪奇緣 雙人組，要價 900 元，付錢後，服務生發現今天套餐大特價，只要 750 元，因此 服務生應該退還 150 元給這三個人，但是這位服務生一時鬼迷心竅，決定暗槓 60 元，只退了 90 元給這三個遊客。
那麼:三人各出 300 元 - 服務生還給他們一人 30 元 = 三人各出 270 元。270 元× 3人+服務生私吞的60元= 810 + 60 = 870 !?怎麼不是900元呢?還 有 30 元去哪了呢?請用敘述的方式，儘量清楚解釋問題出在哪裡。

實際上公司收到的錢是：750元

第一個部分驗證
三人付出去的錢總額是810元
扣除服務生私吞的60元
最終等於公司收到的750元

問題在於：付出去的總額跟服務生暗槓的錢加總是沒有意義的
因為服務生暗槓的60元是從這三個人付出去的錢拿的，並不是原本需要付出的900元，所以如果錯誤加總就會導致這樣奇怪的結果。
